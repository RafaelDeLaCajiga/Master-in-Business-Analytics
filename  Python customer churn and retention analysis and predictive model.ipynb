First off all, the data showed us that customers with a yearly and two year contract plans are more loyal to the company and are less inclined to churn. A direct correlation found was between the length of the contract and churn decision. Since clients with longer contracts (Two and One year contracts) tend to churn less, we see an impact in churn percentage spike up with clients on a Month-to-Month contract. With regards to the type of contract, out of the customers with month-to-month contracts, 1861 have churned, while 2524 have stayed with the company. For one-year contracts, close to 184 clients have churned, while 1469 have not. For two-year contracts, 61 clients have churned out of 1879. 

Overall, the data analysis highlights the importance of contract length and price points, while gender, credit score (to mention a couple) are not significant factors that directly impact churn. ```python


```python
#import data 
Customer_data = pd.read_csv('./A2.csv', index_col=0)

#function for random choice sub services 
def fill_na_phone(row):
    if row['PhoneService'] == 1: 
        return np.random.choice(['Yes', 'No'])
    elif row['PhoneService'] == 0: 
        return 'No'

def fill_na_internet(row):
    if row['Has_InternetService'] == 1: 
        return np.random.choice(['Yes', 'No'])
    elif row['Has_InternetService'] == 0: 
        return 'No'

# confusion matrix for models 

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("""
Normalized confusion matrix""")
#    else:
#        print("""
#Confusion matrix, without normalization""")
    
    
np.random.seed(78)#7>#42>#59
random_int = np.random.randint(0, 10)

    
#clean data by replacing NaN with random entries where entries are missing and ensure all values are 1 and 0 
Customer_data['InternetService'] = Customer_data['InternetService'].fillna(pd.Series(np.random.choice(['DSL', 'Fiber optic', 'No'], size=len(Customer_data))))
Customer_data['Has_InternetService'] = Customer_data['InternetService'].map({'DSL': 1, 'Fiber optic': 1, 'No': 0})
Customer_data['Partner'] = Customer_data['Partner'].fillna(pd.Series(np.random.choice(['Yes', 'No'], size=len(Customer_data)))).map({'Yes': 1, 'No': 0})
Customer_data['PhoneService'] = Customer_data['PhoneService'].fillna(pd.Series(np.random.choice(['Yes', 'No'], size=len(Customer_data)))).map({'Yes': 1, 'No': 0})
Customer_data['Churn'] = Customer_data['Churn'].fillna(pd.Series(np.random.choice(['Yes', 'No'], size=len(Customer_data)))).map({'Yes': 1, 'No': 0})
Customer_data['Female'] = Customer_data['gender'].fillna(pd.Series(np.random.choice(['Female', 'Male'], size=len(Customer_data)))).map({'Female': 1, 'Male': 0})
Customer_data['MultipleLines'] = Customer_data.apply(lambda row: row['MultipleLines'] if pd.notnull(row['MultipleLines']) else fill_na_phone(row), axis=1).map({'Yes': 1, 'No': 0, 'No phone service': 0})
Customer_data['OnlineSecurity'] = Customer_data.apply(lambda row: row['OnlineSecurity'] if pd.notnull(row['OnlineSecurity']) else fill_na_phone(row), axis=1).map({'Yes': 1, 'No': 0, 'No internet service': 0})
Customer_data['OnlineBackup'] = Customer_data.apply(lambda row: row['OnlineBackup'] if pd.notnull(row['OnlineBackup']) else fill_na_phone(row), axis=1).map({'Yes': 1, 'No': 0, 'No internet service': 0})
Customer_data['DeviceProtection'] = Customer_data.apply(lambda row: row['DeviceProtection'] if pd.notnull(row['DeviceProtection']) else fill_na_phone(row), axis=1).map({'Yes': 1, 'No': 0, 'No internet service': 0})
Customer_data['TechSupport'] = Customer_data.apply(lambda row: row['TechSupport'] if pd.notnull(row['TechSupport']) else fill_na_phone(row), axis=1).map({'Yes': 1, 'No': 0, 'No internet service': 0})
Customer_data['StreamingTV'] = Customer_data.apply(lambda row: row['StreamingTV'] if pd.notnull(row['StreamingTV']) else fill_na_phone(row), axis=1).map({'Yes': 1, 'No': 0, 'No internet service': 0})
Customer_data['StreamingMovies'] = Customer_data.apply(lambda row: row['StreamingMovies'] if pd.notnull(row['StreamingMovies']) else fill_na_phone(row), axis=1).map({'Yes': 1, 'No': 0, 'No internet service': 0})
Customer_data['DSL']=Customer_data['InternetService'].replace({'DSL':1, 'Fiber optic': 0, 'No':0}).astype(int)
Customer_data['Fiber']=Customer_data['InternetService'].replace({'DSL':0, 'Fiber optic': 1, 'No':0}).astype(int)
Customer_data['Dependents'] = Customer_data['Dependents'].fillna(pd.Series(np.random.choice(['Yes', 'No'], size=len(Customer_data)))).map({'Yes': 1, 'No': 0}).astype(int)
Customer_data['TotalCharges'] = Customer_data['TotalCharges'].replace(' ', np.nan)
Customer_data['TotalCharges'] = Customer_data['TotalCharges'].astype(float)
Customer_data['Bank_transfer']=Customer_data['PaymentMethod'].replace({"Bank transfer (automatic)":1, 'Electronic check': 0, "Credit card (automatic)":0, 'Mailed check': 0}).astype(int)
Customer_data['Electronic_check']=Customer_data['PaymentMethod'].replace({"Bank transfer (automatic)":0, 'Electronic check': 1, "Credit card (automatic)":0, 'Mailed check': 0}).astype(int)
Customer_data['Credit_Card']=Customer_data['PaymentMethod'].replace({"Bank transfer (automatic)":0, 'Electronic check': 0, "Credit card (automatic)":1, 'Mailed check': 0}).astype(int)
Customer_data['Mailed_Check']=Customer_data['PaymentMethod'].replace({"Bank transfer (automatic)":0, 'Electronic check': 0, "Credit card (automatic)":0, 'Mailed check': 1}).astype(int)

#empty contract lines with mode
Customer_data['Contract'] = Customer_data['Contract'].fillna('Month-to-month')

#empty tenure with median 
Customer_data['tenure'] = Customer_data['tenure'].fillna(29)

#empty Senior Citizen with mode
Customer_data['SeniorCitizen'] = Customer_data['SeniorCitizen'].fillna(0)

#empty CreditScore with median
Customer_data['CreditScore'] = (Customer_data['CreditScore'].fillna(Customer_data['CreditScore'].median()))

#empty EstimatedSalary with median 
Customer_data['EstimatedSalary'] = (Customer_data['EstimatedSalary'].fillna(Customer_data['EstimatedSalary'].median()))

#empty Geography with mode
Customer_data['Geography'] = Customer_data['Geography'].fillna('France')

#drop monthy and TotalCharges empty 
Customer_data = Customer_data.dropna(subset=['TotalCharges']) 
Customer_data = Customer_data.dropna(subset=['MonthlyCharges']) 

#drop outliers of CreditScore
    # Identify the outliers
#q1 = Customer_data['CreditScore'].quantile(0.25)
#q3 = Customer_data['CreditScore'].quantile(0.75)
#iqr = q3 - q1
#upper_limit = q3 + 1.5 * iqr
#lower_limit = q1 - 1.5 * iqr
#outliers = Customer_data[(Customer_data['CreditScore'] > upper_limit) | (Customer_data['CreditScore'] < lower_limit)]
    # Drop the outliers
#Customer_data = Customer_data.drop(outliers.index)

#Created Bins to distinguish customers with Only phone service, Only Internet Service or having Borth Services
Customer_data['BothServices']= (Customer_data['Has_InternetService'] == 1) & (Customer_data['PhoneService'] == 1)
Customer_data['BothServices'] = Customer_data['BothServices'].astype(int)
Customer_data['OnlyPhone'] = (Customer_data['PhoneService']== 1) & (Customer_data['Has_InternetService'] == 0)
Customer_data['OnlyPhone'] = Customer_data['OnlyPhone'].astype(int)
Customer_data['OnlyInternet'] = (Customer_data['PhoneService']== 0) & (Customer_data['Has_InternetService'] == 1)
Customer_data['OnlyInternet'] = Customer_data['OnlyInternet'].astype(int)

#Created Bins depending on the tipe of contract the clients have
Customer_data['Two_year']=Customer_data['Contract'].replace({'Two year':1, 'Month-to-month': 0, 'One year':0}).astype(int)
Customer_data['Month-to-month']=Customer_data['Contract'].replace({'Two year':0, 'Month-to-month': 1, 'One year':0,}).astype(int)
Customer_data['One_year']=Customer_data['Contract'].replace({'Two year':0, 'Month-to-month': 0, 'One year':1,}).astype(int)

#Created bins for each country in the data ser
Customer_data['Germany']=Customer_data['Geography'].replace({'Germany':1, 'France': 0, 'Spain':0}).astype(int)
Customer_data['France']=Customer_data['Geography'].replace({'Germany':0, 'France': 1, 'Spain':0}).astype(int)
Customer_data['Spain']=Customer_data['Geography'].replace({'Germany':0, 'France': 0, 'Spain':1}).astype(int)

#Bins for salary turned into levels of income for customer 
Customer_data['Income_bins'] = Customer_data['EstimatedSalary']
salaries = Customer_data['EstimatedSalary']
Customer_data['Income_bins'] = pd.qcut(salaries, 3, labels=['low_income', 'medium_income', 'high_income'])
Customer_data['High_income']=Customer_data['Income_bins'].replace({'high_income':1, 'low_income':0, 'medium_income':0}).astype(int)
Customer_data['Medium_income']=Customer_data['Income_bins'].replace({'high_income':0, 'low_income':0, 'medium_income':1}).astype(int)

#Bins to evaluate the credit score report for each client. Dividing the clients into 3 different sets of credit outcome.
Customer_data['Credit_bins'] = Customer_data['CreditScore']
credit_scores = Customer_data['CreditScore']
Customer_data['Credit_bins'] = pd.qcut(credit_scores, 3, labels=['Bad_Credit_Score', 'Medium_Credit_Score', 'Good_Credit_Score'])
Customer_data['Good_Credit_Score']=Customer_data['Credit_bins'].replace({'Bad_Credit_Score':0, 'Medium_Credit_Score':0, 'Good_Credit_Score':1}).astype(int)
Customer_data['Medium_Credit_Score']=Customer_data['Credit_bins'].replace({'Bad_Credit_Score':0, 'Medium_Credit_Score':1, 'Good_Credit_Score':0}).astype(int)
Customer_data['Bad_Credit_Score']=Customer_data['Credit_bins'].replace({'Bad_Credit_Score':1, 'Medium_Credit_Score':0, 'Good_Credit_Score':0}).astype(int)

#Divided the monthly charges in triples and created bins to arrange them as high, medium and low
Customer_data['MonthlyChargesBins'] = Customer_data['MonthlyCharges']
MontlyChargesBins = Customer_data['MonthlyCharges']
Customer_data['MonthlyChargesBins'] = pd.qcut(MontlyChargesBins, 3, labels=['HighMonthlyCharge', 'MediumMonthlyCharge', 'LowMonthlyCharge'])
Customer_data['HighMonthlyCharge']= Customer_data['MonthlyChargesBins'].replace({'HighMonthlyCharge':1, 'MediumMonthlyCharge':0, 'LowMonthlyCharge':0}).astype(int)
Customer_data['MediumMonthlyCharge']= Customer_data['MonthlyChargesBins'].replace({'HighMonthlyCharge':0, 'MediumMonthlyCharge':1, 'LowMonthlyCharge':0}).astype(int)



#Month 2 Month and High Monthly Charge
Customer_data['High_Monthly_Charge_M2M']= (Customer_data['Month-to-month'] == 1) & (Customer_data['HighMonthlyCharge'] == 1)
Customer_data['High_Monthly_Charge_M2M'] = Customer_data['High_Monthly_Charge_M2M'].astype(int)



Customer_data['Low_income']=Customer_data['Income_bins'].replace({'high_income':0, 'low_income':1, 'medium_income':0}).astype(int)


#Income And Credit Score
#High Income
Customer_data['High_Income_Good_cs']= (Customer_data['High_income'] == 1) & (Customer_data['Good_Credit_Score'] == 1)
Customer_data['High_Income_Good_cs'] = Customer_data['High_Income_Good_cs'].astype(int)
Customer_data['High_Income_Medium_cs']= (Customer_data['High_income'] == 1) & (Customer_data['Medium_Credit_Score'] == 1)
Customer_data['High_Income_Medium_cs'] = Customer_data['High_Income_Medium_cs'].astype(int)
Customer_data['High_Income_Bad_cs']= (Customer_data['High_income'] == 1) & (Customer_data['Bad_Credit_Score'] == 1)
Customer_data['High_Income_Bad_cs'] = Customer_data['High_Income_Bad_cs'].astype(int)

#Medium Income
Customer_data['Medium_Income_Good_cs']= (Customer_data['Medium_income'] == 1) & (Customer_data['Good_Credit_Score'] == 1)
Customer_data['Medium_Income_Good_cs'] = Customer_data['Medium_Income_Good_cs'].astype(int)
Customer_data['Medium_Income_Medium_cs']= (Customer_data['Medium_income'] == 1) & (Customer_data['Medium_Credit_Score'] == 1)
Customer_data['Medium_Income_Medium_cs'] = Customer_data['Medium_Income_Medium_cs'].astype(int)
Customer_data['Medium_Income_Bad_cs']= (Customer_data['Medium_income'] == 1) & (Customer_data['Bad_Credit_Score'] == 1)
Customer_data['Medium_Income_Bad_cs'] = Customer_data['Medium_Income_Bad_cs'].astype(int)

#Low Income
Customer_data['Low_Income_Good_cs']= (Customer_data['Low_income'] == 1) & (Customer_data['Good_Credit_Score'] == 1)
Customer_data['Low_Income_Good_cs'] = Customer_data['Low_Income_Good_cs'].astype(int)
Customer_data['Low_Income_Medium_cs']= (Customer_data['Low_income'] == 1) & (Customer_data['Medium_Credit_Score'] == 1)
Customer_data['Low_Income_Medium_cs'] = Customer_data['Low_Income_Medium_cs'].astype(int)
Customer_data['Low_Income_Bad_cs']= (Customer_data['Low_income'] == 1) & (Customer_data['Bad_Credit_Score'] == 1)
Customer_data['Low_Income_Bad_cs'] = Customer_data['Low_Income_Bad_cs'].astype(int)

#Drop unnecesary bins and columns
Customer_data = Customer_data.drop(['Income_bins', 'Credit_bins'], axis=1)
Customer_data = Customer_data.drop('MonthlyChargesBins',axis=1)

Customer_data = Customer_data.drop(['gender', 'Charge', 'customerID', 'Surname', 'PaperlessBilling', 'InternetService', 'Contract', 'Geography', 'PaymentMethod'], axis=1)
Customer_data.info()
```

Let's start the analysis by looking at all the clients and see what is our Churn Proportion:

### Churn Frequency and Percentage

```python
# Assuming 'Churn' is a column in the Customer_data dataframe

# Create the countplot
ax = sns.countplot(data=Customer_data, x='Churn')

# Get the total number of observations
total_obs = len(Customer_data)

# Loop through each bar in the plot
for p in ax.patches:
    # Get the height and width of the bar
    height = p.get_height()
    width = p.get_width()
    
    # Calculate the percentage of observations for this category
    percentage = height / total_obs * 100
    
    # Add the percentage and count labels on top of the bar
    ax.annotate('{:.1f}%\n({})'.format(percentage, int(height)), 
                xy=(p.get_x() + width / 2, height), 
                xytext=(0, 5), 
                textcoords='offset points', 
                ha='center', 
                va='bottom', 
                fontsize=8)

# Add a title to the plot
plt.title('Churn Count')

# Show the plot
plt.suptitle('Churned vs Retained', fontsize=12)
plt.title('(Percentage and Count)',fontsize=9)

plt.show()

print(f'\n\nTotal count Churned and Retained = {total_obs}')
```

## Model 

```python
# x-variables
x_var3 = [#'SeniorCitizen', 
#'Partner',
'tenure',
'PhoneService',
'MultipleLines',
#'OnlineSecurity',
#'OnlineBackup',
#'DeviceProtection',
'TechSupport',
#'StreamingTV',
#'StreamingMovies',
'TotalCharges',
#'Churn',
'CreditScore',
'EstimatedSalary',
'MonthlyCharges',
#'Dependents',
'Has_InternetService',
#'Female',
'DSL',
'Fiber',
#'Bank_transfer',
'Electronic_check',
'Credit_Card',
'Mailed_Check',
#'BothServices',
#'OnlyPhone',
#'OnlyInternet',
'Two_year',
'Month-to-month',
'One_year',
#'Germany',
#'France',
#'Spain',
'High_income',
'Medium_income',
'Low_income',
'Good_Credit_Score',
'Medium_Credit_Score',
'Bad_Credit_Score',
'HighMonthlyCharge',
'MediumMonthlyCharge',
'High_Monthly_Charge_M2M',

'High_Income_Good_cs',
'High_Income_Medium_cs',
'High_Income_Bad_cs',
'Medium_Income_Good_cs',
'Medium_Income_Medium_cs',
'Medium_Income_Bad_cs',
'Low_Income_Good_cs',
'Low_Income_Medium_cs',
'Low_Income_Bad_cs'
        ]
y_data3 = Customer_data.loc[ : , 'Churn']
x_data3 = Customer_data.loc[ : , x_var3]

```

```python
# GBM Tuned
# Specify explanatory variables
X = Customer_data.loc[:, x_var3 ]
y = pd.Series(Customer_data.loc[:, 'Churn'])
# Save column names
column_names = X.columns

# Instantiate scaler
scaler = RobustScaler()

# Fit scaler
scaler.fit(X.values)

# Transform explanatory variables
X_scaled = scaler.transform(X.values)


# Train-test split with stratification
X_train, X_test, y_train, y_test = train_test_split(
            X.values,
            y,
            test_size    = 0.25,
            random_state = 219,
            stratify     = y)    


# Oversample
######################################
os = SMOTE(random_state=0)

os_data_x,os_data_y = os.fit_resample(X_train, y_train)
os_data_x = pd.DataFrame(data = os_data_x,
                         columns = x_var3 )
os_data_y= pd.Series(data=os_data_y)

n_cross_sell = len(os_data_y[os_data_y==0])
n_no_cross_sell = len(os_data_y[os_data_y==0])
p_cross_sell = len(os_data_y[
    os_data_y==0])/len(os_data_x)
p_no_cross_sell = len(os_data_y[
    os_data_y==1])/len(os_data_x)


# declaring a hyperparameter space
learn_space        = np.arange(0.1, 3.0, 0.1)
estimator_space    = np.arange(120, 150, 10)
depth_space        = np.arange(3, 5, 1)
warm_start_space   = [True, False]

#min_split_space    = np.arange(2, 400, 100)
#max_features_space = ['auto', 'sqrt', 'log2']
#loss_space         = ['deviance', 'exponential']
#criterion_space    = ['friedman_mse', 'mse', 'mae']

# creating a hyperparameter grid
param_grid = {'learning_rate'     : learn_space,
              'n_estimators'      : estimator_space,
              'max_depth'         : depth_space,
#              'min_samples_split' : min_split_space,
#              'max_features'      : max_features_space,
#              'loss'              : loss_space,
#              'criterion'         : criterion_space,
              'warm_start'        : warm_start_space}

# INSTANTIATING the model object without hyperparameters
gbm_grid = GradientBoostingClassifier(random_state = 219)


# GridSearchCV object
gbm_cv = RandomizedSearchCV(estimator          = gbm_grid,
                           param_distributions = param_grid,
                           cv                  = 3,
                           n_iter              = 10,
                           random_state        = 219,
                           scoring             = make_scorer(roc_auc_score,
                                                 needs_threshold = False))


# FITTING to the FULL DATASET (due to cross-validation)
gbm_cv.fit(X.values, y)


##############################################################################
# Create tuned model

# Instantiate a logistic regression model
gbm_tuned = gbm_cv.best_estimator_

# FIT step is needed as we are not using .best_estimator
gbm_tuned = gbm_tuned.fit(os_data_x.values, os_data_y)


# PREDICTING based on the testing set
gbm_tuned_pred = gbm_tuned.predict(X_test)


# Instantiate scores
gbm_tuned_train_score = gbm_tuned.score(os_data_x.values, os_data_y).round(4)
gbm_tuned_test_score  = gbm_tuned.score(X_test, y_test).round(4)
gbm_tuned_test_gap    = abs(gbm_tuned_test_score - gbm_tuned_train_score).round(4)
gbm_tuned_auc_score   = roc_auc_score(y_true  = y_test, y_score = gbm_tuned_pred).round(4)
gbm_tuned_report      = classification_report(y_test, 
                                       gbm_tuned_pred,
                                       target_names = ['Cross Sell Failed (0)',
                                                       'Cross Sell Successful (1)'])



##############################################################################
# Plot Confusion Matrix

# Unpack confusion matrix
gbm_tuned_tn, \
gbm_tuned_fp, \
gbm_tuned_fn, \
gbm_tuned_tp = confusion_matrix(y_true = y_test, y_pred = gbm_tuned_pred).ravel()

# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, gbm_tuned_pred, labels=[1,0])
np.set_printoptions(precision = 2)



# Assess performance by plotting non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, 
                      classes   = ['Cross-sell = 1','Cross-sell = 0'],
                      normalize = False,  
                      title     = 'Confusion Matrix')
plt.show()

precision = gbm_tuned_tp / (gbm_tuned_tp + gbm_tuned_fp)
recall = gbm_tuned_tp / (gbm_tuned_tp + gbm_tuned_fn)
f1_score_gbm_tuned = 2 * (precision * recall) / (precision + recall)

# Score results
print(f"""
=================== MODEL SUMMARY ===================

Model Type: GBM Tuned

Model Size: {X.shape[1] + 1}

---------------------------------

Tuned GBM F1-score : {f1_score_gbm_tuned}

Tuned GBM AUC Score      : {gbm_tuned_auc_score}

---------------------------------

""")
print(cnf_matrix)
print(classification_report(y_test, gbm_tuned_pred))


##############################################################################
# Plot ROC & AUC

# Calculate the FPR and TPR for all thresholds of the classification
probs = gbm_tuned.predict_proba(X_test)
preds = probs[:,1]
fpr, tpr, threshold = metrics.roc_curve(y_test, gbm_tuned_pred)
roc_auc = metrics.auc(fpr, tpr)

# Plot ROC
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()


# Source of code:
# https://stackoverflow.com/questions/25009284/how-to-plot-roc-curve-in-python
```

## Recommendation

According to the features of importance graph, we are able to determine that the presence of fiber, the possibility of paying via electronic check and the tech support, represents, in our model, the three most important extra services that the company must focus on. 
In other words, these services have a direct impact on the company's capacity of retention. Favoring these services and using them as leverage might be the key to diminishing the churn tendency of certain customers. 

Concerned about the churn of our clients, we established that the two main factors of this tendency lie in the type of contract and the contracts' charges. Indeed, the month-to-month contract represents the contract with the highest percentage of departure (due to its easiness to leave), and we noticed that the churn tendency is accentuated for monthly charges above $60. 

As a solution to this situation, we recommend driving the customers away from month-to-moth contracts and to redirect them to longer contract. To lure them we will need to implement exclusive benefits in one year and two-year contract. 

One-year contract subscribers will benefit from every month-to-month advantages. 
In addition, a faster fiber-optic cables facilities, the possibility of paying via Bank transfer, the last device protection system, and the possibility of having multiple lines for a lower cost. 

Two-years contracts will benefit from all the one-year contract advantages plus the possibility of paying via electronic check and mailed check, discount for switching from DSL to fiber, possibility to add new mobile line for free if internet subscription plan is taken, a better tech support 24/7 and finally a discount for any new mobile lines taken if there are dependents. 

Finally, month-to-month subscription will have less advantages. They will benefit from a basic tech support with limited hours of availability, internet service, streaming movies and tv, and credit card payment only. 
Because month-to-month buyers are mostly people with lower income, introducing a reward referral system will be an opportunity to increase this consumer's segment, but keeping in mind that this type of contract has the highest clients' turnover. 
Our reward referral system would work such as the client who is going to refer a friend will get renewal credit for data based on the plan that this friend picks with us.  

The strategy behind this is to increase the quantity of month-to-month subscriber in order to, in a second time, redirect them to our longer contract plan, attractive because of their benefits. 

Even if one-year and two-years contract originally aimed businesses and home clients, with higher purchasing power and willingness to be loyal in exchange of dedicated services, the chances of having individuals with lower income still need to be envisaged. 

To conclude, we would finally recommend keeping monthly charges up to a certain level (around $60) for basic offers such as month to month contracts and longer contracts. 
Financially speaking, margin of month-to-month customer will cover benefits given to the two-years and one-year contracts.
