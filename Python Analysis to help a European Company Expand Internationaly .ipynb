## Problem 1

```python
Market = dataco.groupby(['Market'])['Sales'].sum().round(0).astype(int)
sales_by_year = Market.to_frame()
sales_by_year = sales_by_year.reset_index()
total_row = pd.DataFrame({
    'Market': ['Total'],
    'Sales': [sales_by_year['Sales'].sum()]
})
sales_by_year = sales_by_year.append(total_row, ignore_index=True)
sales_by_year = sales_by_year.sort_values("Sales", ascending=False) 
sales_by_year
```

```python
Countries = dataco.groupby(['Market','Order Country'])['Sales'].sum().round(0).astype(int)
sales_by_year = Countries.to_frame()
sales_by_year = sales_by_year.reset_index()
total_row = pd.DataFrame({
    'Order Country': ['Total'],
    'Sales': [sales_by_year['Sales'].sum()]
})
overall_sales = sales_by_year.append(total_row, ignore_index=True)
overall_sales = overall_sales.sort_values("Sales", ascending=False) 
overall_sales
```

```python
latam_sales = overall_sales[overall_sales['Market'] == 'LATAM']
total_row = pd.DataFrame({
    'Market': ['Total'],
    'Sales': [overall_sales['Sales'].sum()]
})
latam_sales = latam_sales.append(total_row, ignore_index=True)
latam_sales = latam_sales.sort_values("Sales", ascending=False) 
latam_sales
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,10)})
sns.set_style("white")

# Filter out the Total row
latam_sales = latam_sales.loc[latam_sales['Market'] != 'Total']

# Filter the dataframe to show only the top 10 countries
top_10_countries = latam_sales.nlargest(10, 'Sales')

# Create the barplot
ax = sns.barplot(x="Order Country", y="Sales", hue="Order Country", data=top_10_countries, palette="colorblind")

# Adjust the width of the bars
for patch in ax.patches:
    current_width = patch.get_width()
    diff = current_width - 0.4
    patch.set_width(0.4)
    patch.set_x(patch.get_x() + diff * .5)

# Set the x-axis label and title
ax.set(xlabel='Market', ylabel='Sales in 000', title='LATAM Sales by Country')

# Move the legend to the top right corner
ax.legend(loc='upper right')

# Rotate the x-axis labels
plt.xticks(rotation=45)

plt.show()
```

```python
mexico_data = cost_of_living[cost_of_living['Country'] == 'Mexico']
brazil_data = cost_of_living[cost_of_living['Country'] == 'Brazil']
mexico_data
brazil_data
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data
labels = ['Cost of Living', 'Rent', 'Cost of Living plus Rent', 'Groceries', 'Restaurant', 'LPPI']
mexico_values = [mexico_data['Cost of Living Index'].values[0], mexico_data['Rent Index'].values[0], mexico_data['Cost of Living Plus Rent Index'].values[0], mexico_data['Groceries Index'].values[0], mexico_data['Restaurant Price Index'].values[0], mexico_data['Local Purchasing Power Index'].values[0]]
brazil_values = [brazil_data['Cost of Living Index'].values[0], brazil_data['Rent Index'].values[0], brazil_data['Cost of Living Plus Rent Index'].values[0], brazil_data['Groceries Index'].values[0], brazil_data['Restaurant Price Index'].values[0], brazil_data['Local Purchasing Power Index'].values[0]]

# Create the figure and axis objects
fig, ax = plt.subplots(figsize=(8, 6))

# Define the bar width and opacity
bar_width = 0.35
opacity = 0.8

# Plot the bars for Brazil and Mexico
mexico_bars = ax.bar(labels, mexico_values, bar_width, alpha=opacity, color='g', label='Mexico')
brazil_bars = ax.bar(labels, brazil_values, bar_width, alpha=opacity, color='y', label='Brazil')


# Find the highest value for each country and its index
brazil_max_value = max(brazil_values)
brazil_max_index = brazil_values.index(brazil_max_value)
mexico_max_value = max(mexico_values)
mexico_max_index = mexico_values.index(mexico_max_value)

# Add the value of the highest bar behind the second highest bar for each country
if brazil_max_value > mexico_max_value:
    ax.text(brazil_max_index, mexico_max_value * 1.05, f'{brazil_max_value:.1f}', ha='center', va='bottom', fontsize=10)
else:
    ax.text(mexico_max_index, brazil_max_value * 1.05, f'{mexico_max_value:.1f}', ha='center', va='bottom', fontsize=10)

# Set the x tick positions and labels
xtick_positions = np.arange(len(labels))
ax.set_xticks(xtick_positions + bar_width/2)
ax.set_xticklabels(labels)

# Set the x axis label rotation
plt.xticks(rotation=45)

# Add the x axis dividing lines
for x in xtick_positions:
    ax.axvline(x=x+bar_width/2, color='gray', alpha=1)

# Set the x and y axis labels and title
ax.set_xlabel('Index')
ax.set_ylabel('Value')
ax.set_title('Cost of Living and Rent Comparison: Mexico vs. Brazil')
```
